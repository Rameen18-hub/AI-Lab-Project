```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
import pickle

df = pd.read_csv("processed_data.csv")


X = df.drop("GPA", axis=1)   
y = df["GPA"]               


train_X, test_X, train_y, test_y = train_test_split(
    X, y, test_size=0.3, shuffle=True, random_state=42
)

scaler = StandardScaler()

train_X_scaled = scaler.fit_transform(train_X)
test_X_scaled  = scaler.transform(test_X)

model = SVR(kernel='rbf', C=10, gamma='scale')
model.fit(train_X_scaled, train_y)

data_to_save = {
    "model": model,
    "scaler": scaler,
    "columns": X.columns.tolist()
}

with open("best_regression_model.pkl", "wb") as f:
    pickle.dump(data_to_save, f)

print("Model + Scaler Saved Successfully!")
```

<div class="alert alert-block alert-info">
    6. <b>Applying Classifier / Model Training</b>
</div>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import pickle

df = pd.read_csv('processed_data.csv')


X = df.drop('GradeClass', axis=1)   
y = df['GradeClass']                


train_X, test_X, train_y, test_y = train_test_split(
    X, y, test_size=0.3, shuffle=True, random_state=42
)

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
     random_state=42
)
rf_model.fit(train_X, train_y)
rf_pred = rf_model.predict(test_X)
rf_acc = accuracy_score(test_y, rf_pred)
print(f"RandomForest Accuracy: {rf_acc*100:.2f}%")

gb_model = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
gb_model.fit(train_X, train_y)
gb_pred = gb_model.predict(test_X)
gb_acc = accuracy_score(test_y, gb_pred)
print(f"GradientBoosting Accuracy: {gb_acc*100:.2f}%")

best_model = rf_model if rf_acc >= gb_acc else gb_model
with open('best_classification_model.pkl', 'wb') as file:
    pickle.dump(best_model, file)

print("âœ… Best model saved as 'best_classification_model.pkl'")
```
